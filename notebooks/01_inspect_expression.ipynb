```python
import pandas as pd 
path = "../data/raw/TCGA.LUAD.sampleMap_HiSeqV2.gz"

expr_head = pd.read_csv(
    path,
    sep="\t",
    compression="gzip",
    nrows = 5,
)

expr_head
```

```python
sample_ids = expr_head.columns[1:]

def sample_type(sample_id: str) -> str:
    code = sample_id.split("-")[-1]
    
    if code == "01":
        return "tumor"
    elif code == "11":
        return "normal"
    else:
        return "other"

labels = pd.DataFrame({
    "sample_id": sample_ids,
    "sample_type": [sample_type(s) for s in sample_ids],
})

labels.head()
```
